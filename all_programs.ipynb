```python
# présent de l'indicatif
class VerbePremierGroupe():
    def __init__(self, verb):
        if verb[-2:] != "er":
            raise ValueError("Le verbe n'est pas du premier groupe !")
        self.verb = verb
    def conjugue(self, temps = "present_indicatif"):
        if temps == "present_indicatif":
            return self.present_indicatif()
    def present_indicatif(self):
        radical = self.verb[:-2].lower()
        radical_nous = radical + "e" if radical[-1] == "g" else radical
        return [
            f"Je {radical}e",
            f"Tu {radical}es",
            f"Il {radical}e",
            f"Nous {radical_nous}ons",
            f"Vous {radical}ez",
            f"Ils {radical}ent",
        ]


def present():
    verb = input("Taper votre verbe du premier groupe (q pour quitter) : ")
    #if verb == "q":
    #    sys.exit(0)
    try:
        mon_verbe = VerbePremierGroupe(verb)
        for v in mon_verbe.conjugue():
            print(v, "\n")
    except ValueError as e:
        print(e, "\n")
    #present()
present()
```

```python
# distance entre deux mots
def compute_hamming(word1, word2):
    if len(word1) != len(word2):
        raise ValueError("Les deux mots ne sont pas de la même taille")
    diff = 0
    for i in range(len(word1)):
        if word1[i].lower() != word2[i].lower():
            diff = diff + 1
    return diff 

def word_distance():
    try:
        print("La distance de hamming est : ", compute_hamming(input("Votre premier mot : "), input("Votre second mot : ")))
    except ValueError as e:
        print(e, "\n")
    #if input("Taper q pour quitter ou entrée pour continuer : ") == "q":
    #    exit()
    #word_distance()
word_distance()
```

```python
# Inverse d'un mot
def reverse_word(word):
    word = list(word.lower())
    word.reverse()
    word[0] = word[0].upper()
    return "".join(word)

def reverse_logic():
    word = input("Taper le mot à renverser (q pour quitter) : ")
    #if word == "q":
    #    exit()
    print(reverse_word(word))
    #reverse_logic()
reverse_logic()
```

```python
import math
import re
def solve(a,b,delta, power = 1):
    return (-b + power * math.sqrt(delta))/(2 * a)

def compute_delta(a, b, c):
    return math.pow(b,2) - 4 * (a * c)

def soluceIsReal(delta):
    return delta > 0

def polynome():
    print("Afin de calculer les solutions de votre polynome, veuillez entrer dans l'ordre.\n")
    try:
        polynome = input("Taper votre polynome du second degrés (ex: 2x²+3x-5) : ")
        m = re.match("(?:([-.0-9]+)x[²|^2]{1,2})*[+]*(?:([-.0-9]+)x)*[+]*([-.0-9]+)*", polynome)
        if m != None:
            a = float(m.group(1) if m.group(1) != None else 0)
            b = float(m.group(2) if m.group(2) != None else 0)
            c = float(m.group(3) if m.group(3) != None else 0)
            delta = compute_delta(a,b,c)
            if a != 0 and soluceIsReal(delta):
                x = [ solve(a,b,delta), solve(a, b, delta, -1) ]
                print(f"x1 : {x[0]}\nx2 : {x[1]}")
            else:
                raise ValueError("Votre polynome ne peut pas être résolu dans le monde réel.")
        else:
            raise ValueError("Vous n'avez pas taper un polynome de type ax²+bx+c !")
        
    except ValueError as e:
        print(e)
polynome()
```

```python
# agent secret
def encrypt(word = ""):
    bytes = [b + 3 if b + 3 < 91 else ((b + 3) % 90) + 64 for b in bytearray(word, 'utf-8')]
    return bytearray(bytes).decode('utf-8')

def decrypt(word = ""):
    bytes = [b - 3 if b - 3 > 64 else 90 - (b % 65) for b in bytearray(word, 'utf-8')]
    return bytearray(bytes).decode('utf-8')

def agent_secret_name():
    print("Merci de choisir ce que vous voulez faire : \n")
    print("    [1] Encrypter le nom d'un agent\n")
    print("    [2] Décrypter le nom d'un agent\n")
    try:
        choice = input("Que voulez-vous faire (q pour quitter) ? ")
        #if choice == "q":
        #    exit()
        choice = int(choice)
        if choice == 1:
            word = input("Taper le nom de l'agent ultra secret : ")
            print(encrypt(word.upper()))
        elif choice == 2:
            word = input("Taper le nom encrypté de l'agent ultra secret : ")
            print(decrypt(word.upper()))
        else:
            raise ValueError("")
    except ValueError:
        print("Veuillez taper un chiffre correspondant à une des options du menu.\n")
    #agent_secret_name()

print("Bienvenue dans le programme ultra secret d'anonymisation de votre nom d'agent !\n")
print("Nous utilisons une méthode ultra secrète pour encrypter et décrypter les agents ultra secrets de l'organisation.\n")
agent_secret_name()
```

```python
# balises html
import requests

ALLOW_ATTRIBUTES = False

def get_html_page(url):
    r = requests.request("GET", url)
    if r.status_code != 200:
        raise ValueError(f"Impossible de se connecter au site web (Erreur {r.status_code})")
    return r.content

def display_html_tags():
    url = input("Taper l'addresse du site web à crawler (q pour quitter) : ")
    #if url == "q":
    #    exit()
    try:
        raw = get_html_page(url)
        is_html_tag = False
        temp_html_tag = ""
        for b in raw:
            if b == 60: # < = 60 and > = 62
                is_html_tag = True
                temp_html_tag = ""
                continue
            if b == 62:
                print(temp_html_tag)
                is_html_tag = False
            if is_html_tag and not ALLOW_ATTRIBUTES and b == 32:
                is_html_tag = False
            if is_html_tag:
                temp_html_tag += chr(b)
    except ValueError as e:
        print(e, "\n")
    #display_html_tags()
display_html_tags()
```

```python
# month calendar
import datetime

WEEKDAYS = {
    "LUN": 0,
    "MAR": 1,
    "MER": 2,
    "JEU": 3,
    "VEN": 4,
    "SAM": 5,
    "DIM": 6
}

def get_first_day_of_month(month, year):
    return datetime.date(year, month, 1)

def get_last_day_of_month(month, year):
    next_month = datetime.date(year if month < 12 else year + 1, month % 12 + 1, 1)
    return next_month + datetime.timedelta(-1)

def print_calendar_header():
    for weekday in WEEKDAYS:
        print(weekday, end="")
        print("  ", end="")

def show_calendar(month, year):
    start_date = get_first_day_of_month(month, year)
    end_date = get_last_day_of_month(month, year)
    current_date = start_date
    print_calendar_header()
    print("\n", end="")
    weekday = 0
    while current_date <= end_date:
        if weekday == 0:
            print("\n", end="")
        if current_date.weekday() == weekday:
            print("" if current_date.day > 9 else " ", current_date.day, " ", end="", sep="")
            current_date += datetime.timedelta(1)
        else:
            print("   ", end="")
        print("  ", end="")
        weekday = (weekday + 1) % 7
    print("\n")

def calendar():
    try:
        date = input("Rentrer la date du jour pour lequel afficher le calendrier (ex: 23/01/2006) ou q pour quitter : ")
        #if date == "q":
        #    exit()
        if date == "":
            date = datetime.date.today().strftime("%d/%m/%Y")
        date = date.split("/")
        if len(date) != 3:
            raise ValueError("Vous n'avez pas saisi la date au bon format !")
        date = datetime.date(int(date[2]), int(date[1]), int(date[0]))
        show_calendar(date.month, date.year)
    except ValueError as e:
        print(e)
    #calendar()
calendar()
```

```python
import random

class PileFaceObject():
    def __init__(self, turn, is_pile = True):
        self.is_pile = is_pile
        self.turn = turn

class PileFaceGame():
    def __init__(self):
        self.reset()

    def launch_coin(self):
        return random.randint(0, 1) == 1

    def reset(self):
        self.__games = []
        self.__pile_indexes = []
        self.__face_indexes = []

    def play_games(self, nb_game = 1, log = False):
        for i in range(nb_game):
            result = self.launch_coin()
            if log:
                print(f"Tour #{i} : Vous obtenez", "'Pile'" if result else "'Face'")
            self.__games.append(PileFaceObject(i, result))
            if result:
                self.__pile_indexes.append(len(self.__games) - 1)
            else:
                self.__face_indexes.append(len(self.__games) - 1)
    
    def nb_pile(self):
        #return len(list(filter(lambda x: x.is_pile, self.__games)))
        # len is a no cost operation whereas filter does loop in the background
        return len(self.__pile_indexes)

    def nb_face(self):
        return len(list(filter(lambda x: not x.is_pile, self.__games)))
    
    def total_games(self):
        return len(self.__games)

def play(game):
    try:
        n = int(input("Veuillez entrer le nombre de jeu que l'on va simuler (0 pour quitter) : "))
        #if n == 0:
        #    exit()
        game.reset()
        game.play_games(n)
        print("Toutes les parties sont terminés !")
        print("Voici les résultats :")
        print(f"Nombre de Pile : {game.nb_pile()} ({game.nb_pile()/game.total_games() * 100 :.0f} %)")
        print(f"Nombre total de tirages :", game.total_games())
    except ValueError:
        print("Vous n'avez pas entré un nombre entier positif !")
    #play(game)
        


game = PileFaceGame()
print("Bienvenue dans la simulation du jeu de Pile ou Face !")
play(game)
```
